1. **Introduction**:
   - Seaborn is a statistical data visualization library in Python.
   - It provides a high-level interface for drawing attractive and informative statistical graphics.
   - It is built on top of Matplotlib and integrates closely with Pandas data structures.

2. **Key Features**:
   - **High-Level Interface**: Seaborn provides a simple and intuitive API for creating complex statistical visualizations.
   - **Statistical Estimation**: Seaborn offers functions to summarize and visualize distributions and relationships in datasets.
   - **Integration with Pandas**: Seaborn seamlessly works with Pandas DataFrames, making it easy to visualize data stored in tabular formats.
   - **Attractive Default Styles**: Seaborn comes with visually pleasing default styles and color palettes that enhance the aesthetics of plots.
   - **Flexible Plotting Functions**: Seaborn offers a wide range of plotting functions for various types of data visualization, including univariate and bivariate plots, categorical plots, and time series plots.
   - **Advanced Customization**: Seaborn allows for extensive customization of plots, including control over colors, styles, labels, and annotations.

3. **Commonly Used Plotting Functions**:
   - **Distribution Plots**: `sns.histplot()`, `sns.kdeplot()`, `sns.distplot()` for visualizing univariate distributions.
   - **Categorical Plots**: `sns.barplot()`, `sns.countplot()`, `sns.boxplot()` for visualizing relationships between categorical variables.
   - **Scatter Plots**: `sns.scatterplot()`, `sns.jointplot()`, `sns.pairplot()` for visualizing relationships between numerical variables.
   - **Regression Plots**: `sns.regplot()`, `sns.lmplot()`, `sns.residplot()` for visualizing linear regression models.
   - **Matrix Plots**: `sns.heatmap()`, `sns.clustermap()` for visualizing matrices of data.

4. **Integration with Matplotlib**:
   - Seaborn is built on top of Matplotlib and can be used alongside Matplotlib functions for additional customization.
   - Matplotlib functions can be invoked directly for fine-grained control over plot elements.

5. **Installation and Compatibility**:
   - Seaborn can be installed via pip (`pip install seaborn`) or conda (`conda install seaborn`).
   - It is compatible with Python 3.x and integrates well with other scientific computing libraries such as NumPy, Pandas, and SciPy.

6. **Documentation and Resources**:
   - Seaborn provides comprehensive documentation with examples and tutorials to help users get started.
   - The Seaborn community is active, with forums and online resources available for support and collaboration.

Overall, Seaborn is a powerful and versatile tool for creating informative and visually appealing statistical visualizations in Python, making it an essential library for data analysis and exploration tasks.

## Features:

- total_bill: Total bill amount (including tax and tip).
- tip: Tip amount given by the customer.
- sex: Gender of the customer (male or female).
- smoker: Whether the customer is a smoker (yes or no).
- day: Day of the week (Thursday, Friday, Saturday, Sunday).
- time: Time of the day (Lunch or Dinner).
- size: Number of people in the dining party.

```python
# Relational plot 

sns.scatterplot(x=df.total_bill,y=df.tip)

# It is a axis level function

# Let say if you are in matplotlib then u have to tell the pc that what is x label and y label 
```

```python
# Replot -> Figure level function -> Square shaped 
# seaborn recommends regplot but we can use scatterplot instead for ease now
# what seaborn suggests is we should use figure level functions more 

sns.relplot(x=df.total_bill,y=df.tip, kind='scatter')
plt.show()
```

```python
#Hue parameter 
# particular features

sns.scatterplot(x=df.total_bill,y=df.tip, hue=df.sex)
```

```python
# style Parameter 
# we can add one more layer of information, lets say i want see yaha kitne log dinner ke time aaye or kitne lunch ke time aaye 
```

```python
#Now we have 4 layers of information
sns.scatterplot(x=df.total_bill,y=df.tip, hue=df.sex, style=df.time)
```

```python
# size parameter 
# Lets say i want to check koi customer kitne logo ke saath aya
```

```python
sns.scatterplot(x=df.total_bill,y=df.tip, hue=df.sex, style=df.time, size=df.size)
```

```python
# Line plot 
# Line plot is basically a scatterplot 
# this is where we use continuity in graph 
# Let say we have scatterplot points, if we join all these points we will get a line plot 
# it is more used for time-series data 
```

Time series data refers to a sequence of data points collected or recorded at successive and equally spaced intervals of time. In other words, it is a series of observations or measurements obtained over a specific period, where each data point is associated with a particular timestamp or time period.

Time series data is just a fancy way of saying data that changes over time. Imagine you're keeping track of something, like the temperature outside, every hour. Each hour, you write down the temperature. That's a time series! It's like a timeline of data points where each point has a time stamp.

The Gapminder dataset typically includes the following columns:

- country: The name of the country.
- year: The year for which the data was recorded.
- pop: Population of the country.
- continent: The continent to which the country belongs.
- lifeExp: Life expectancy at birth for the country.
- gdpPercap: GDP per capita (income per person) for the country.
- iso_alpha: ISO 3166-1 alpha-3 country code.

These columns provide information about various socio-economic indicators for different countries over time, allowing for analysis of trends in global development, comparisons between countries, and exploration of relationships between different variables.
